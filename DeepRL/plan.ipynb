我觉得如果是没有RL基础的话，直接上大部头不是最优的选择。
1. OpenAI spinning up是一个很精简的文档，可以按着他的路线来学习基础的知识。David silver的视频作辅助，（Cousera最近 U Alberta 也出了RL基础的教学视频，看起来安排地很有逻辑）sutton的introduction也作为参考资料。
2. OpenAI的spining up过了一遍后，其实就可以直接看DQN DDPG这样的经典文章了，看看这些算法是怎么提出来的，RL有怎样的威力，它们的motivation是怎样的，当遇到不懂的，又回过头来看看sutton 的introduction里传统RL算法。
3. 走到这一步，如果有ml dl的基础的话，（没有也可）（CS285和david视频一起看也可）就直接开始看Sergey Levine的 CS285 (原294-112)，这门课基本涵盖了DRL前沿的大部分内容介绍，前几章是基础算法，后几章是拓展开的RL研究的分支，（但也不是所有）做完这些，就开始读paper啦
